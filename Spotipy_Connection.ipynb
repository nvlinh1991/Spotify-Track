### Connection and queyry 1000 tracks in 2021 (csv is then locally saved)

```python
my_id = "myid"
my_code = "mycode"
df = SpotifyConnector().data_query(user_id=my_id, user_code=my_code, year=2021)
```

#### There are 2 values in track_id are duplicated.
#### Some track_id has the same track_name, which could be from the same album. 
#### For simplicity, duplicated track_id will be removed

```python
df.drop_duplicates(subset='track_id', keep='first', inplace=True)
```

## Go further with some basic preprocessing

```python
# Call Objects from function
preproc_tool = PreProc(data=df.copy())
preproc_tool.variable_exploration()
```

#### Around 30% of tracks have high popularity (>=70). 60% are insignificant (popularity < 10)
#### J Balvin accounts for 10% of all tracks
#### ~1/4 track names contains "NEW YEAR"

### There could be some cleaning needed to remove stopwords, punctuation

```python
clean_data = preproc_tool.text_cleaning()
```

```python
clean_df = clean_data['DataFrame']
new_column = clean_data['Cleaning Columns']
```

### Data after cleaning text with new columns (stored in list new_column)

```python
clean_df.head(5)
```

### POPULAR TRACK
### Let's have a look characteristics of popular track. Select threshold for popularity score >= 80 (around 15% of total tracks)

```python
high_threshold = 80
popular_track = clean_df[clean_df.popularity >= high_threshold].reset_index(drop=True)
```

```python
popular_track.shape
```

```python
popular_track.describe()
```

```python
popular_track.head(5)
```

```python
preproc_tool = PreProc(data=popular_track.copy())
preproc_tool.word_cloud(list_column=['clean_track_name', 'clean_artist_name'], max_word=10)
```

### Top popular tracks are mostly coming from duets (feat), remix
### Those tracks are from Olivia Rodrigo, Tyler The Creator, and Justin Bieber
